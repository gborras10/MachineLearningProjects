## Train tree classifier
First, we fit a :class:`~sklearn.tree.DecisionTreeClassifier` using the
:func:`~sklearn.datasets.load_iris` dataset.

```python
# Import a data set.
dataset = load_iris()

# Consider only two features. 
indices_features = [0, 1]
X = dataset.data[:, indices_features]
y = dataset.target

# Initialize an instance of a Decision Tree classifier

random_state = 10 # for reproducibility
max_leaf_nodes = 5
classifier = DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes, 
                                    random_state=0)

# Train the classifier using the training data 
classifier.fit(X, y)
```

```python
fig, ax = plt.subplots(figsize=(14, 14))  # whatever size you want
tree.plot_tree(classifier, fontsize=14, ax=ax)
plt.show()
```

```python
# Plot decision regisons

decision_function = classifier.predict 

ax = pltu.plot_2D_decision_regions(X, y, 
                                   decision_function=decision_function,
                                   decision_levels='auto',
                                   cmap=plt.cm.rainbow) 

fontsize = 14    

_ = ax.set_title('Decision tree classification',
                 fontsize=fontsize)
             
_ = ax.set_xlabel(dataset.feature_names[indices_features[0]], 
                  fontsize=fontsize)
_ = ax.set_ylabel(dataset.feature_names[indices_features[1]], 
                  fontsize=fontsize)
```
